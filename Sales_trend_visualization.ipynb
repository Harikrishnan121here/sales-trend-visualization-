```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("colorado_motor_vehicle_sales.csv")

# Show first rows
print(df.head())

# Check missing values
print(df.isnull().sum())

# -----------------------------------
# SALES TREND VISUALIZATION
# -----------------------------------

# Group yearly sales
yearly_sales = df.groupby('year')['sales'].sum()

# Line Chart
plt.figure(figsize=(10,5))
yearly_sales.plot(kind='line', marker='o')

plt.title("Yearly Sales Trend")
plt.xlabel("Year")
plt.ylabel("Sales")

plt.grid(True)
plt.show()

# -----------------------------------
# COUNTY WISE SALES
# -----------------------------------

county_sales = df.groupby('county')['sales'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(12,6))
county_sales.plot(kind='bar')

plt.title("Top 10 County Sales")
plt.xlabel("County")
plt.ylabel("Sales")

plt.show()

# -----------------------------------
# QUARTER WISE SALES
# -----------------------------------

quarter_sales = df.groupby('quarter')['sales'].sum()

plt.figure(figsize=(7,7))
quarter_sales.plot(kind='pie', autopct='%1.1f%%')

plt.title("Quarter Wise Sales Distribution")
plt.ylabel("")

plt.show()
```

## Insights

- Vehicle sales showed fluctuations across different years.
- Some counties contributed significantly higher sales.
- Quarter-wise sales indicate seasonal demand patterns.
